```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating   = [2,3,4,3,2]
working  = [7,8,7,2,2]
playing  = [8,5,7,8,13]

plt.plot([],[],label="Sleeping",linewidth=5,color="r")
plt.plot([],[],label="Eating",linewidth=5,color="g")
plt.plot([],[],label="Working",linewidth=5,color="b")
plt.plot([],[],label="Playing",linewidth=5,color="k")

plt.stackplot(days, sleeping,eating,working,playing, colors=["r","g","b","k"] )

plt.xlabel("X")
plt.ylabel("Y")
plt.title("stackplot example")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

slices = [7,2,2,13] 
act = ["sleeping","eating","working","playing"]
cols = ["red","blue","Green","cyan"]

plt.pie(slices,labels=act,colors = cols,startangle = 90,explode = (0,0.1,0,0),shadow=True,autopct="%.1f")
plt.title("pie chart")
plt.legend()
plt.show()
```

```python
import datetime as dt
from matplotlib import style
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure
style.use('ggplot')

start = dt.datetime(2018,1,1) 
end   = dt.datetime(2018,2,2)

df = web.DataReader('TSLA','iex',start,end)

dates = df.index
plt.scatter(df.index,df.open,label='open',color='blue', marker="*")
plt.scatter(df.index,df.close,label='close',color='red', marker="*")
plt.scatter(df.index,df.low,label='low',color='green', marker="*")
plt.scatter(df.index,df.high,label='high',color='black', marker="+")

#plt.bar(df.index,df.volume,label='Volume',color='cyan')

figure(num="None",figsize=(30,30))
plt.xticks(rotation=90,fontsize = 2)                ## rotate the x labels by 90* to increase visibility

plt.grid(True ,color="g")                         ## gives out horizontal lines
plt.xlabel("$ Value")
plt.ylabel(" Duration ")
plt.title("TSLA Share Performance")
plt.legend()
plt.savefig('index.png')
plt.show()
```
